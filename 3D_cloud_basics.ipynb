# Simple point cloud data 3D visual

Here demonstarted a simple 3D array data of point cloud(X, Y, Z) coordinate and showed how to vizualize it in 3D using matplot.

```python
# Required Library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Sample LiDAR point cloud data (x, y, z coordinates in meters)
lidar_data = np.array([
    [1.0, 2.0, 0.0],
    [1.5, 2.5, 0.1],
    [2.0, 3.0, 0.2],
    [2.5, 3.5, 0.3],
    [3.0, 4.0, 0.4],
    [1.0, 1.0, 0.0],
    [1.5, 1.5, 0.1],
    [2.0, 2.0, 0.2],
    [2.5, 2.5, 0.3],
    [3.0, 3.0, 0.4],
])
```

```python
# Created a Pandas DataFrame from the LiDAR data
df = pd.DataFrame(lidar_data, columns=['x(Meters)', 'y(Meters)', 'z(Meters)'])
df
```

```python
#Created a empty 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

# OUTPUT

```python
#Extract x, y, and z coordinates
x_coords = df['x(Meters)'].values
y_coords = df['y(Meters)'].values
z_coords = df['z(Meters)'].values

# Print or use the extracted coordinates
print("X Coordinates:", x_coords)
print("Y Coordinates:", y_coords)
print("Z Coordinates:", z_coords)

# Created a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_coords, y_coords, z_coords)

# Set labels and title
ax.set_xlabel('X-axis (meters)')
ax.set_ylabel('Y-axis (meters)')
ax.set_zlabel('Z-axis (meters)')
ax.set_title('3D LiDAR Point Cloud')

plt.show()
```
